## Stacks

A Stack is an ordered collection of elements where items are added and removed from one end, commonly referred to as the "top". The ordering principle is referred to as LIFO (Last in, first out)



# Challenge 1: Push and Pop

Instructions:

1. Create an empty stack.
2. Push the values "A", "B", and "C" onto the stack in that order.
3. Pop the top value and store it in a variable called top.
4. Print the entire stack and print the value that was removed.
5. Pop the next value from the stack and print it.
6. Push the value "D" onto the stack.

```python
stack = []

stack.append("A")
stack.append("B")
stack.append("C")

top = stack.pop()

print("Stack:", stack)
print("Removed:", top)

print("Removed:", stack.pop())

stack.append("D")

print("Removed:", stack.pop())

print("Final Stack:", stack)
```

```python
# Simplified version (relying on built-in functions)

class Stack:
    def __init__(self):
        self.items = []

    def push(self, elements):
        self.items.append(elements)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)

    def is_empty(self):
        return self.items == []
    
    #f

    def __str__(self):
        print(self.items)
```

```python
stack_obj = Stack()

stack_obj.push("A")
stack_obj.push("B")
stack_obj.push("C")

top = stack_obj.pop()

print(stack_obj.items)
print(top)

print(stack_obj.peek())
print(stack_obj.size())
```

```python
# Custom Stack class (from scratch)

class CustomStack():

    class Node:
        def __init__(self, data):
            self.data = data
            self.below = None

    def __init__(self):
        self.top = None

    def push(self, value):
        if not self.top:
            self.top = self.Node(value)

        else:
            new_node = self.Node(value)
            new_node.below = self.top
            self.top = new_node

    # Assignment 2: Build the logic for the pop and is_empty functions

    def pop(self):
        if self.top:
            removed = self.top.data
            self.top = self.top.below
            return removed

    def is_empty(self):
        return self.top == None
```

```python
stack = CustomStack()

stack.push("A")
stack.push("B")
stack.push("C")

print(stack.pop())

print(stack.is_empty())
```
