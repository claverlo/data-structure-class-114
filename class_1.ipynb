```python
# Variables
my_name = "Leo"
age = 100
is_active = True
print(f"My name is {my_name}")
```

```python
# Collections (Lists, Dictionaries, Tuples, Sets)
students = ["Leomar", "Freys", "Sergio", "Charles", "Trishon", "Aymen"]

diff = ["String", 10, True, 'c']

my_dictionary = {
    "name":"Leo",
    "age":125,
    "is_active":True
}

print(students[0])
print(my_dictionary["name"])
```

```python
# Expressions - Operators
a = 10
b = 3

# Arithmetic
print(a + b)
print(a - b)
print(a * b)
print(a / b)

# Relational
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
print(a == b)
```

```python
# Functions
def my_function():
    a = 10
    b = 3
    print(a + b)

def my_function_with_parameters(a,b):
    print(a + b)

def my_function_with_return():
    a = 10
    b = 3
    return a+b

my_function()
my_function_with_parameters(10,3)

result = my_function_with_return()
print(result)
```
